```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# reading/loading the data 
df = pd.read_csv("Experience-Salary.csv")

#first five rows
df.head()
```

```python
# lets change the columns name for convinence
df.rename(columns={"exp(in months)":"experience", 
                   "salary(in thousands)":"salary"},inplace=True)
```

```python
df.info()
```

```python
# some basic idea about the data
df.describe()
```

```python
input = df.iloc[:,:-1]  # independent feature
target = df.iloc[:,-1]  # dependent feature
```

```python
#spliting the dataset into training and testing

X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.3, random_state=42)
```

```python
# linear regression model

model = LinearRegression()
```

```python
# train the model 
model.fit(X_train,y_train)
```

```python
# make prediction from the model
y_pred = model.predict(X_test)
```

```python
# visualizing the best fit line/regression line 

plt.figure(figsize=(12,6))  # size of graph

sns.scatterplot(x=X_train['experience'],y=y_train)  # scatter plot
plt.plot(X_test,y_pred,c='red')   # line plot
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Regression Plot")
plt.show()
```

```python
# evaluating metrics of the model 

print("R-Square error of the model: ", r2_score(y_test,y_pred))
print("Mean absolute error of the model: ", mean_absolute_error(y_test,y_pred))
print("Mean Squared error of the model: ", mean_squared_error(y_test,y_pred))
```
